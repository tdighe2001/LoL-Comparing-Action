```python
#creating df with just split, datacompleteness, and side where datacompleteness, and side are bool columns. 
#complete, Blue = True; partial, Red = False.
missdf = df[['split', 'datacompleteness', 'side']]
missdf = missdf.replace(r'^\s*$', np.nan, regex=True)
missdf = missdf.replace({'datacompleteness' : { 'complete' : True, 'partial' : False}})
missdf = missdf.replace({'side' : { 'Blue' : True, 'Red' : False}})
missdf['split_missing'] = missdf['split'].isna()
missdf.head()
```

```python
datacomp_dist = (
    missdf
    .assign(split_missing=missdf['split'].isna())
    .pivot_table(index='datacompleteness', columns='split_missing', aggfunc='size')
)

# Added just to make the resulting pivot table easier to read.
datacomp_dist.columns = ['split_missing = False', 'split_missing = True']

datacomp_dist = datacomp_dist / datacomp_dist.sum()
datacomp_dist
#The two columns look very different, which is evidence that 'split''s missingness does depend on 'datacompleteness'
```

```python
#plotting above observation on bar graph
datacomp_dist.plot(kind='barh', title='Datacompleteness by Missingness of split', barmode='group')
```

```python
side_dist = (
    missdf
    .assign(split_missing=missdf['split'].isna())
    .pivot_table(index='side', columns='split_missing', aggfunc='size')
)

# Added just to make the resulting pivot table easier to read.
side_dist.columns = ['split_missing = False', 'split_missing = True']

side_dist = side_dist / side_dist.sum()
side_dist
```

```python
#plotting above observation on bar graph
side_dist.plot(kind='barh', title='Side by Missingness of Split', barmode='group')
```

```python
#IMPLEMENTATION OF PERM TEST
import scipy.stats as stats
def mcar_permutation_tests(missdf,n_permutations):

    israel = missdf.copy()
    split_missing = israel[israel['split'].isnull()]
    split_not_missing = israel[~israel['split'].isnull()]

    datacomp_missing = split_missing['datacompleteness'].mean()
    datacomp_not_missing = split_not_missing['datacompleteness'].mean()
    
    ssmean_missing = split_missing['side'].mean()
    ssmean_not_missing = split_not_missing['side'].mean()
    
    datacomptest_statistic = abs(datacomp_missing - datacomp_not_missing)
    sstest_statistic = abs(ssmean_missing - ssmean_not_missing)
    
    datacomppermutation_test_statistics = []
    sspermutation_test_statistics = []
    
    for i in range(n_permutations):
    # Shuffle the 'datacompleteness' values while keeping the 'split' and 'side' values fixed
        datacompshuffled = israel['datacompleteness'].sample(replace=True, frac=1).reset_index(drop=True)
        israel['datacompleteness'] = datacompshuffled
        ssshuffled = israel['side'].sample(replace=True, frac=1).reset_index(drop=True)
        israel['side'] = ssshuffled
    
    # Split the shuffled data into two groups - rows where 'split' was missing and rows where 'split' was not missing
        split_missing_shuffled = israel[israel['split'].isnull()]
        split_not_missing_shuffled = israel[~israel['split'].isnull()]
    
    # Calculate the means of the 'datacompleteness' column for each group
        datacomp_mean_missing_shuffled = split_missing_shuffled['datacompleteness'].mean()
        datacomp_mean_not_missing_shuffled = split_not_missing_shuffled['datacompleteness'].mean()
        ss_mean_missing_shuffled = split_missing_shuffled['side'].mean()
        ss_mean_not_missing_shuffled = split_not_missing_shuffled['side'].mean()
    
    # Calculate the test statistic (the absolute difference in group means) for the permuted data
        datacomppermutation_test_statistic = abs(datacomp_mean_missing_shuffled - datacomp_mean_not_missing_shuffled)
        sspermutation_test_statistic = abs(ss_mean_missing_shuffled - ss_mean_not_missing_shuffled)
    
    # Add the test statistic to the list of test statistics from each permutation
        datacomppermutation_test_statistics.append(datacomppermutation_test_statistic)
        sspermutation_test_statistics.append(sspermutation_test_statistic)

    # Calculate the p-value
    datacompp_value = (np.sum(datacomppermutation_test_statistics >= datacomptest_statistic)) / (n_permutations)
    ssp_value = (np.sum(sspermutation_test_statistics >= sstest_statistic)) / (n_permutations)

    return datacompp_value,ssp_value
```

```python
mcar_permutation_tests(missdf,500)
```

```python
# Null: All tier 1 leagues have the same amount of action per game.
# Alternative: The tier 1 league VCS is significantly more action-packed than other regions
# Test statistic: difference in group means for kills
# Permutation testing, difference in the distributions of VCS and non VCS tier 1 leagues
testing = games[['is_vcs','team kpm']]
observed = testing.groupby('is_vcs').mean().diff().iloc[-1][0]
observed
```

```python
n = 500
differences = []
shuffled = testing.copy()
for _ in range(n):
    shuffled['shuffled tkpm'] = np.random.permutation(shuffled['team kpm'])
    diff = shuffled.groupby('is_vcs').mean().diff().iloc[-1]['shuffled tkpm']
    differences.append(diff)
```

```python
(differences >= observed).mean()
```

```python
fig = px.histogram(pd.DataFrame({'difference in group means':differences}),x = 'difference in group means', histnorm = 'probability')
fig.add_vline(x=observed, line_color='red')
fig.update_layout(xaxis_range=[-0.1, 0.3])
fig.show()
```
